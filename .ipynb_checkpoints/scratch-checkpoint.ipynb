```python
import scanpy as sc
import anndata as ad
import numpy as np
```

```python
sdata = sc.read_h5ad('submission.h5ad')
```

```python
# Define our path to the training anndata
tr_adata_path = "vcc_data/adata_Training.h5ad"
print(1)

# Read in the anndata
tr_adata = ad.read_h5ad(tr_adata_path)
print(2)

# Filter for non-targeting
ntc_adata = tr_adata[tr_adata.obs["target_gene"] == "non-targeting"]
print(3)

cap = 200000-len(sdata.obs['target_gene'])-1
ntc_adata = ntc_adata[:cap]
print(4)

ntc_adata.X = np.log1p(ntc_adata.X)
```

```python
sdata = ad.concat(
        [
            sdata,
            ntc_adata,
        ]
)
```

```python
sdata.write('FINAL.h5ad')
```
